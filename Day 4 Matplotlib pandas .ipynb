```python
import matplotlib.pyplot as plt
# Plot histograms for all columns in the DataFrame
plt.figure(figsize=(10, 6))
for column in input.columns:
    input['Age'].plot(kind='hist', bins=20, alpha=0.2, label=column)
plt.title('Histograms of Age')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#box plot for specific attribute
hist = input['Smokes'].hist(bins=50)
plt.title('Histogram for Smoke Column')
plt.xlabel('No of Patients taken smoke')
plt.ylabel('Frequency')
plt.show()
```

```python
#plot specific attribute
input.plot(y='Age', figsize=(9,6))
```
